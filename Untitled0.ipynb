```python
import kagglehub
import os

# Download dataset
dataset_path = kagglehub.dataset_download("huynhangia/trainingdata")
print("Dataset path:", dataset_path)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import os

# Diagnose the issue by printing the contents of the downloaded dataset directory
print("Contents of dataset_path:", os.listdir(dataset_path))

# Further diagnose by printing the contents of the 'archive' directory
archive_path = os.path.join(dataset_path, "archive")
print("Contents of archive_path:", os.listdir(archive_path))

# Corrected path based on the output of os.listdir(archive_path)
# The dataset seems to have classes directly inside the 'archive' folder
train_dir = os.path.join(dataset_path, "archive")

classes = [c for c in os.listdir(train_dir)
           if os.path.isdir(os.path.join(train_dir, c))]
num_classes = len(classes)

print("Classes:", classes)
print("Number of classes:", num_classes)
```

```python
import matplotlib.pyplot as plt

# Đếm số ảnh trong mỗi class
class_counts = {}

for cls in classes:
    class_dir = os.path.join(train_dir, cls)
    num_images = len([
        f for f in os.listdir(class_dir)
        if f.lower().endswith(('.png', '.jpg', '.jpeg'))
    ])
    class_counts[cls] = num_images

# In ra để kiểm tra
print("Image count per class:")
for cls, count in class_counts.items():
    print(f"{cls}: {count}")
plt.figure(figsize=(8, 5))
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel("Waste Class")
plt.ylabel("Number of Images")
plt.title("Distribution of Images per Waste Class")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import shutil
import random

SOURCE_DIR = train_dir            # chính là dataset_path/archive
TARGET_DIR = "dataset_split"      # thư mục mới để train

TRAIN_RATIO = 0.8
VAL_RATIO   = 0.1
TEST_RATIO  = 0.1

random.seed(42)

print("Splitting dataset...")
for cls in classes:
    src_cls_dir = os.path.join(SOURCE_DIR, cls)
    images = [f for f in os.listdir(src_cls_dir)
              if os.path.isfile(os.path.join(src_cls_dir, f))]

    random.shuffle(images)

    n_total = len(images)
    n_train = int(n_total * TRAIN_RATIO)
    n_val   = int(n_total * VAL_RATIO)

    split_map = {
        "train": images[:n_train],
        "val":   images[n_train:n_train + n_val],
        "test":  images[n_train + n_val:]
    }

    for split, img_list in split_map.items():
        dst_dir = os.path.join(TARGET_DIR, split, cls)
        os.makedirs(dst_dir, exist_ok=True)

        for img in img_list:
            shutil.copy(
                os.path.join(src_cls_dir, img),
                os.path.join(dst_dir, img)
            )

print("Dataset split completed!")
```

```python
#cell thống kê data sau khi phân tách
from collections import Counter
import matplotlib.pyplot as plt
import os

def count_images_per_class(base_dir):
    class_counts = {}
    for cls in sorted(os.listdir(base_dir)):
        cls_path = os.path.join(base_dir, cls)
        if os.path.isdir(cls_path):
            class_counts[cls] = len(os.listdir(cls_path))
    return class_counts

splits = ["train", "val", "test"]
data = {}

for split in splits:
    split_dir = os.path.join("dataset_split", split)
    data[split] = count_images_per_class(split_dir)
    print(f"\n{split.upper()} distribution:")
    for k, v in data[split].items():
        print(f"{k}: {v}")
```

```python
#cell vẽ biểu đồ sau khi chia
plt.figure(figsize=(14,4))

for i, split in enumerate(splits):
    plt.subplot(1,3,i+1)
    plt.bar(data[split].keys(), data[split].values())
    plt.title(f"{split.upper()} distribution")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32   # nếu bị OOM → giảm còn 16
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation"
)
```

```python
import numpy as np
from sklearn.utils.class_weight import compute_class_weight

# Lấy label của tập TRAIN
labels = train_generator.classes

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(labels),
    y=labels
)

class_weight_dict = dict(enumerate(class_weights))

print("Class weights:")
for cls_idx, weight in class_weight_dict.items():
    print(f"Class {cls_idx}: {weight:.2f}")
```

```python
base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

base_model.trainable = False  # Phase 1
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation="relu")(x)
x = Dropout(0.4)(x)
outputs = Dense(num_classes, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=outputs)
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

```

```python
# Cell 5: TRAIN PHASE 1 (CHỈ TRAIN HEAD, CÓ class_weight)
callbacks = [
    EarlyStopping(
        monitor="val_loss",
        patience=7,
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=4
    )
]

history_phase1 = model.fit(
    train_generator,
    epochs=15,
    validation_data=val_generator,
    class_weight=class_weight_dict,   # mât cái này là chết
    callbacks=callbacks
)
```

```python
# Cell 6: FINE-TUNING (Optimized)
base_model.trainable = True

for layer in base_model.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history_phase2 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=15,
    class_weight=class_weight_dict,
    callbacks=callbacks
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
# 1. Reset generator để đảm bảo thứ tự dữ liệu
val_generator.reset()

# 2. Dự đoán trên tập validation
pred_probs = model.predict(val_generator, verbose=1)
y_pred = np.argmax(pred_probs, axis=1)

# 3. Nhãn thật
y_true = val_generator.classes

# 4. Tên class theo đúng thứ tự
class_names = list(val_generator.class_indices.keys())

print("Class indices mapping:")
print(val_generator.class_indices)
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names
)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix – Waste Classification")
plt.tight_layout()
plt.show()
```

```python
#cell7
acc = history_phase1.history["accuracy"] + history_phase2.history["accuracy"]
val_acc = history_phase1.history["val_accuracy"] + history_phase2.history["val_accuracy"]

plt.figure(figsize=(8,5))
plt.plot(acc, label="Train Accuracy")
plt.plot(val_acc, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Đường dẫn đến dataset (từ code gốc)
# data_dir = 'archive'
# Corrected path to use the previously defined train_dir
data_dir = train_dir

# Kích thước ảnh và batch size
img_height, img_width = 224, 224  # MobileNetV2 yêu cầu
batch_size = 32

# Data augmentation cho train
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Split 80% train, 20% val
)

# Generator cho train
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Generator cho validation
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Số lớp
num_classes = len(train_generator.class_indices)
print("Classes:", train_generator.class_indices)
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Predict trên validation set
Y_pred = model.predict(validation_generator)
y_pred = np.argmax(Y_pred, axis=1)

# True labels
true_labels = validation_generator.classes

# Classification report
target_names = list(train_generator.class_indices.keys())
print(classification_report(true_labels, y_pred, target_names=target_names))

# Nếu muốn confusion matrix
print(confusion_matrix(true_labels, y_pred))
```

```python
import matplotlib.pyplot as plt

# Combine history from both phases for loss and accuracy
loss_combined = history_phase1.history['loss'] + history_phase2.history['loss']
val_loss_combined = history_phase1.history['val_loss'] + history_phase2.history['val_loss']
acc_combined = history_phase1.history['accuracy'] + history_phase2.history['accuracy']
val_acc_combined = history_phase1.history['val_accuracy'] + history_phase2.history['val_accuracy']

# Plot loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(loss_combined, label='Train Loss')
plt.plot(val_loss_combined, label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(acc_combined, label='Train Accuracy')
plt.plot(val_acc_combined, label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
from tensorflow.keras.applications import EfficientNetB0, VGG16
from tensorflow.keras.models import Sequential
import time
```

```python
epochs = 15 # Define epochs

# Define VGG16 model
base_model_vgg = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
base_model_vgg.trainable = False  # Freeze

model_vgg = Sequential([
    base_model_vgg,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

# Compile
model_vgg.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train
history_vgg = model_vgg.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size,
    epochs=epochs
)

# Save
model_vgg.save('waste_vgg16.keras')
```

```python
# Define EfficientNetB0 model
base_model_eff = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
base_model_eff.trainable = False  # Freeze

model_eff = Sequential([
    base_model_eff,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

# Compile
model_eff.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train
history_eff = model_eff.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size,
    epochs=epochs
)

# Save
model_eff.save('waste_efficientnetb0.keras')
```

```python
models = {
    'MobileNetV2': model,
    'EfficientNetB0': model_eff,
    'VGG16': model_vgg
}

histories = {
    'MobileNetV2': history_phase2,
    'EfficientNetB0': history_eff,
    'VGG16': history_vgg
}

metrics_dict = {}

for name, m in models.items():
    # Predict
    Y_pred = m.predict(validation_generator)
    y_pred = np.argmax(Y_pred, axis=1)
    true_labels = validation_generator.classes

    # Report
    report = classification_report(true_labels, y_pred, target_names=target_names, output_dict=True)
    metrics_dict[name] = {
        'Accuracy': report['accuracy'],
        'Precision (macro)': report['macro avg']['precision'],
        'Recall (macro)': report['macro avg']['recall'],
        'F1-Score (macro)': report['macro avg']['f1-score']
    }
    print(f"\nClassification Report for {name}:\n")
    print(classification_report(true_labels, y_pred, target_names=target_names))

# Hiển thị bảng so sánh metrics
import pandas as pd
metrics_df = pd.DataFrame(metrics_dict).T
print(metrics_df)
```

```python
histories = {
    'MobileNetV2': history_phase2,
    'EfficientNetB0': history_eff,
    'VGG16': history_vgg
}
for name, h in histories.items():
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(h.history['loss'], label='Train Loss')
    plt.plot(h.history['val_loss'], label='Val Loss')
    plt.title(f'Loss over Epochs - {name}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(h.history['accuracy'], label='Train Accuracy')
    plt.plot(h.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'Accuracy over Epochs - {name}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.show()
```

```python
comparison_dict = {}

for name, m in models.items():
    # Parameters
    params = m.count_params()

    # Model size (file size)
    file_path = f'waste_{name.lower().replace("net", "net")}.keras'  # Adjust tên file
    size_mb = os.path.getsize(file_path) / (1024 * 1024) if os.path.exists(file_path) else 0

    # Inference time (predict 1 batch)
    start_time = time.time()
    _ = m.predict(validation_generator, steps=1)
    inf_time = time.time() - start_time

    comparison_dict[name] = {
        'Total Parameters': params,
        'Model Size (MB)': round(size_mb, 2),
        'Inference Time (s/batch)': round(inf_time, 4)
    }

# Bảng so sánh
comp_df = pd.DataFrame(comparison_dict).T
print(comp_df)

# Kết hợp với metrics
full_comparison = pd.concat([metrics_df, comp_df], axis=1)
print("\nFull Comparison:\n")
print(full_comparison)
```
